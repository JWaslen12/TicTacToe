```python
import random
import numpy as np

def initialize():
    """
    Creates an empty numpy array to serve as the board.
    """
    board = np.zeros([3,3], dtype = int)
    return board

def possibilities(board):
    """
    Finds where the board == 0, ie. where players can place their markers.
    """
    place = np.where(board == 0)
    return place

def place(board, player, location):
    """
    Places a marker for the given player, at a specific location.
    """
    if board[int(location[0])][int(location[1])] ==0:
        board[int(location[0])][int(location[1])] = player
        valid = True
    else:
        valid = False
    print(board)
    return board, valid

def diag_win(board, player):
    """
    Tests if one diagonal is all one player, then flips the board to test for the other diagonal.
    """
    if np.all(np.diag(board)==player) or np.all(np.diag(np.fliplr(board))==player):
        return True
    else:
        return False
    
def row_win(board, player):
    """
    Tests if one row is all one player, then goes through every row.
    """
    if np.all(board[0] == player) or np.all(board[1]==player) or np.all(board[2]==player):
        return True
    else:
        return False
def col_win(board, player):
    """
    Tests if one columns is all one player, then goes through every column.
    """
    if np.all(board[:,0]==player) or np.all(board[:,1]==player) or np.all(board[:,2]==player):
        return True
    else:
        return False
    
def random_place(board):
    """
    Randomly places a marker on possible locations for player two. Only called if only one player.
    """
    marker = possibilities(board)
    if len(marker) > 0:
        index = random.randint(0, len(marker[0])-1)
        marker = str(marker[0][index])+str(marker[1][index])
        board, valid = place(board, 2, marker)
    return board, valid

def evaluate(board, player):
    """
    Evaluates based on conditions of the board.
    """
    winner = 0
    for player in [1, 2]:
        if row_win(board, player) or col_win(board, player) or diag_win(board, player):
            winner = player
    if np.all(board != 0) and winner == 0:
        winner = -1
    return winner

def finish(winner):
    """
    Gives prompts at end of game.
    """
    if winner == -1:
            play_again = input("""Tie Game! Enter 'Y' to play again!""")
    else:
        play_again = input("""Congratulations! Player """+str(winner)+""" is the Winner! Enter 'Y' to play again!""")
    if play_again == 'Y' or play_again == 'y':
        play_game()
        
def play_game():
    """
    Call required functions as needed to play game.
    """
    board = initialize()
    winner = 0
    ans = input('One or Two Players? (input as 1 or 2)')
    if ans == '2':
        two_player = True
    else:
        two_player = False
    if two_player:
        print(board)
        while winner == 0:
            loc1 = input('Player One Placement (input as two digits): ')
            board, valid = place(board, 1, loc1)
            while valid is False:
                print('Not a Valid Placement! Try Again!')
                loc1 = input('Player One Placement (input as two digits): ')
                board, valid = place(board, 1, loc1)
            winner = evaluate(board, 1)
            if winner == 0:
                loc2 = input('Player Two Placement (input as two digits): ')
                board, valid = place(board, 2, loc2)
                while valid is False:
                    print('Not a Valid Placement! Try Again!')
                    loc2 = input('Player Two Placement (input as two digits): ')
                    board, valid = place(board, 2, loc2)
                winner = evaluate(board, 2)
        finish(winner)
    else:
        print(board)
        while winner == 0:
            loc1 = input('Player One Placement (input as two digits): ')
            board, valid = place(board, 1, loc1)
            while valid is False:
                print('Not a Valid Placement! Try Again!')
                loc1 = input('Player One Placement (input as two digits): ')
                board, valid = place(board, 1, loc1)
            winner = evaluate(board, 1)
            if winner == 0:
                board, valid = random_place(board)
                winner = evaluate(board, 2)
        finish(winner)

play_game()
```
